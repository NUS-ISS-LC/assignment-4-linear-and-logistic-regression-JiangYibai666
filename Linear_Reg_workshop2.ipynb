# Linear Regression Example 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# 1) Load the dataset 'OmniPower.csv'
df = pd.read_csv('OmniPower.csv')

# 2) Create a Linear Regression Model with two independent variables (Price and Promotion), one dependent variable (Sale)
X = df[['Price', 'Promotion']]
y = df['Sales']

model = LinearRegression()
model.fit(X, y)

# 3) Provide input values and predict the Sale
print(f"\nPredict examples:")
sample_input = [[59, 200], [79, 400], [99, 600]]
predictions = model.predict(sample_input)

for i, (price, promotion) in enumerate(sample_input):
    print(f"Price: {price}, Promotion: {promotion} -> Predicted Sales: {predictions[i]:.2f}")

# 4) What is the R square value?
y_pred = model.predict(X)
r_squared = r2_score(y, y_pred)
print(f"\nR square value: {r_squared:.4f}")
```
